```python
import yfinance as yf
from datetime import date
import pandas as pd
from google.colab import drive
```

```python
drive.mount('/content/drive')
```

```python
top_tech_companies = yf.Sector('technology').top_companies.index.tolist()
```

```python
historical_data = pd.DataFrame()
financials_data = pd.DataFrame()
balancesheet_data = pd.DataFrame()
cashflow_data = pd.DataFrame()
major_holders_data = pd.DataFrame()
institutional_holders_data = pd.DataFrame()
actions_data = pd.DataFrame()
dividends_data = pd.DataFrame()
splits_data = pd.DataFrame()

for company in top_tech_companies:
  company_data = yf.Ticker(company)

  hist = company_data.history(start='2020-01-01', end=date.today())
  hist['ticker'] = company
  historical_data = pd.concat([historical_data, hist])

  fin = company_data.financials
  fin['ticker'] = company
  financials_data = pd.concat([financials_data, fin])

  bal = company_data.balancesheet
  bal['ticker'] = company
  balancesheet_data = pd.concat([balancesheet_data, bal])

  cash = company_data.cashflow
  cash['ticker'] = company
  cashflow_data = pd.concat([cashflow_data, cash])

  mh = company_data.major_holders
  mh['ticker'] = company
  major_holders_data = pd.concat([major_holders_data, mh])

  ih = company_data.institutional_holders
  ih['ticker'] = company
  institutional_holders_data = pd.concat([institutional_holders_data, ih])

  acts = company_data.actions
  acts['ticker'] = company
  actions_data = pd.concat([actions_data, acts])

  div = company_data.dividends
  div['ticker'] = company
  dividends_data = pd.concat([dividends_data, div])

  split = company_data.splits
  split['ticker'] = company
  splits_data = pd.concat([splits_data, split])
```

```python
historical_data.to_csv("/content/drive/MyDrive/ISAN Project/Datasets/historical_data.csv", index=True)
financials_data.to_csv("/content/drive/MyDrive/ISAN Project/Datasets/financials_data.csv", index=True)
balancesheet_data.to_csv("/content/drive/MyDrive/ISAN Project/Datasets/balancesheet_data.csv", index=True)
cashflow_data.to_csv("/content/drive/MyDrive/ISAN Project/Datasets/cashflow_data.csv", index=True)
major_holders_data.to_csv("/content/drive/MyDrive/ISAN Project/Datasets/major_holders_data.csv", index=True)
institutional_holders_data.to_csv("/content/drive/MyDrive/ISAN Project/Datasets/institutional_holders_data.csv", index=True)
actions_data.to_csv("/content/drive/MyDrive/ISAN Project/Datasets/actions_data.csv", index=True)
dividends_data.to_csv("/content/drive/MyDrive/ISAN Project/Datasets/dividends_data.csv", index=True)
splits_data.to_csv("/content/drive/MyDrive/ISAN Project/Datasets/splits_data.csv", index=True)
```
